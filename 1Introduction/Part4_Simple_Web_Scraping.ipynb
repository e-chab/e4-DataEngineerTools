# Requête HTTP 

Un requête HTTP est une requête basée sur le protocole TCP, elle fait partie de la couche application de la couche OSI. Elle permet d'accéder aux données mise à disposition sur une adresse IP (ou url résolue par un DNS) et un port. 

Les deux ports les plus utilisés dans le web sont le 80 pour les sites en HTTP et le 443 pour les sites en HTTPS. HTTPS est une variable du protocole HTTP basé sur le protocole TLS.

Il existe de nombreux types de requêtes selon la convention `REST`: 
- GET
- POST
- PUT 
- DELETE
- UPDATE.

Dans notre cas, nous allons utiliser la plupart du temps des GET et potentiellement des POST. 
- Le GET permet comme son nom l'indique de récupérer des informations en fonction de certains paramètres. 
- Le POST nécessite un envoi de données pour récupérer des données. Le body du post est, la plupart du temps, envoyé sous la forme d'un objet JSON.

Ces requêtes encapsulent un certain nombre de paramètres qui permettent soient d'identifier une provenance et un utilisateur ou de réaliser différentes actions.

````python
#Exercice 1 : 
class HttpRequester:
    def __init__(self, user_agent=None):
        self.user_agent = user_agent or 'Mozilla/5.0 (Windows NT 6.1; Win64; x64; rv:47.0) Gecko/20100101 Firefox/47.0'

    def get(self, url, timeout=10, retries=3):
        headers = {'User-Agent': self.user_agent}
        try:
            response = requests.get(url, headers=headers, timeout=timeout)
            response.raise_for_status()
            return response
        except Exception as e:
            if retries > 0:
                return self.get(url, timeout, retries-1)
            else:
                print(f"Request failed: {e}")
                return None
````

````python
#Exercice 2 :
import re #le module standard de Python pour la gestion des expressions régulières (regex)
from urllib.parse import urlparse #importer la fonction urlparse du module standard Python urllib.parse
                                  #urlparse sert à analyser (parser) une URL et à en extraire ses différentes parties : le protocole (scheme), le domaine (netloc), le chemin (path), les paramètres, etc.
from bs4 import BeautifulSoup #importer la classe BeautifulSoup du module bs4 (BeautifulSoup4). 
                              #BeautifulSoup une bibliothèque Python qui sert à parser (analyser) et manipuler du code HTML ou XML.
                              #Elle transforme le texte HTML en un objet Python, ce qui facilite la recherche et l’extraction d’éléments (balises, attributs, texte, etc.) dans une page web.

# Supprimer tous les espaces superflus d'une string
def remove_extra_spaces(s):
    return re.sub(r'\s+', ' ', s).strip()

# Nettoyer une string HTML pour la rendre intelligible
def clean_html_string(html):
    soup = BeautifulSoup(html, "html.parser")
    text = soup.get_text()
    text = re.sub(r'[^\w\s.,;:!?\'"-]', '', text)  # Enlever les caractères spéciaux
    return remove_extra_spaces(text)

# Récupérer le domaine d'une URL
def get_domain(url):
    parsed_url = urlparse(url)
    return parsed_url.netloc
````

### Exercice 3

Améliorer la classe développé précédemment.

- Ajouter une méthode pour récupérer l'objet soup d'un url
- Récupérer une liste de User Agent et effectuer une rotation aléatoire sur celui à utiliser
- Utiliser cette classe pour parser une page HTML et récupérer : le titre, tous les H1 (si ils existent), les liens vers les images, les liens sortants vers d'autres sites, et le texte principal.

````python
#Exercice 3

import requests  # Pour faire des requêtes HTTP (télécharger le contenu d'une page web)
import random    # Pour choisir aléatoirement un User-Agent dans une liste
from bs4 import BeautifulSoup  # Pour parser et manipuler le code HTML

class AdvancedHttpRequester:
    def __init__(self, user_agents=None):
        self.user_agents = user_agents or [
            "Mozilla/5.0 (Windows NT 6.1; Win64; x64; rv:47.0) Gecko/20100101 Firefox/47.0",
            "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36",
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
        ]

    def get(self, url, timeout=10, retries=3):
        headers = {'User-Agent': random.choice(self.user_agents)}
        try:
            response = requests.get(url, headers=headers, timeout=timeout)
            response.raise_for_status()
            return response
        except Exception as e:
            if retries > 0:
                return self.get(url, timeout, retries-1)
            else:
                print(f"Request failed: {e}")
                return None

    def get_soup(self, url, timeout=10, retries=3):
        response = self.get(url, timeout, retries)
        if response:
            return BeautifulSoup(response.text, "html.parser")
        return None

    def parse_page(self, url):
        soup = self.get_soup(url)
        if not soup:
            return None

        # Titre de la page
        title = soup.title.string if soup.title else None

        # Tous les H1
        h1_tags = [h1.get_text(strip=True) for h1 in soup.find_all("h1")]


        # Liens vers les images
        images = [img.get("src") for img in soup.find_all("img") if img.get("src")]


        # Liens sortants (vers d'autres sites)
        domain = requests.utils.urlparse(url).netloc
        links = []
        for a in soup.find_all("a", href=True):
            href = a["href"]
            if href.startswith("http") and domain not in href:
                links.append(href)


        # Texte principal
        text = soup.get_text(separator=" ", strip=True)


        return {
            "title": title,
            "h1_tags": h1_tags,
            "images": images,
            "external_links": links,
            "main_text": text
        }

# Exemple d'utilisation :
# requester = AdvancedHttpRequester()
# result = requester.parse_page("https://www.esiee.fr/")
# print(result)
````

# Exploitation des appels d'API



Losque le front du site récupère des données sur une API gérée par le back, un appel d'API est réalisé. Cet appel est recensé dans les appels réseaux. Il est alors possible de re-jouer cet appel pour récupérer à nouveau les données. Il est très facile de récupérer ces appels dans l'onglet Network de la console développeur de Chrome ou FireFox. La console vous permet de copier le code CURL de la requête et vous pouvez ensuite la transformer en code Python depuis le site https://curl.trillworks.com/.

Souvent les APIs sont bloquées avec certains paramètres. L'API vérifie que dans les headers de la requête HTTP ces paramètres sont présents :
* un token généré à la volée avec des protocoles OAuth2 (ou moins développés).
* un referer provenant du site web (la source de la requête), très facile à falsifier.

# Exercice Final  

Exercice Final
Utilisez tout ce que vous avez appris pour récupérer des articles de News avec une catégorie. Il est souvent intéressant de partir des flux RSS pour commencer :

Les données doivent comprendre :
- Le texte important propre
- L'url
- Le domaine
- la catégorie
- Le titre de l'article
- Le titre de la page
- (Facultatif) : les images

Tips : 

- Taper le nom de votre média favoris + RSS (par exemple : https://www.lemonde.fr/rss/)
- Aller dans le DOM de la page 
- Trouver les catégories et les liens vers les articles

````python
# Exercice Final : Récupérer des articles de News via RSS
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse
def clean_text(html):
    soup = BeautifulSoup(html, "html.parser")
    text = soup.get_text()
    return ' '.join(text.split())
def get_rss_articles(rss_url, category=None):
    response = requests.get(rss_url)
    soup = BeautifulSoup(response.content, "xml")
    articles = []
    for item in soup.find_all("item"):
        title = item.title.get_text(strip=True) if item.title else None
        link = item.link.get_text(strip=True) if item.link else None
        domain = urlparse(link).netloc if link else None
        description = item.description.get_text() if item.description else ""
        text = clean_text(description)
        # Catégorie (si présente dans le flux RSS)
        cat = item.category.get_text(strip=True) if item.category else category
        # Images (si présentes)
        enclosure = item.find("enclosure")
        image = enclosure["url"] if enclosure and enclosure.has_attr("url") else None
        articles.append({
            "title": title,
            "url": link,
            "domain": domain,
            "category": cat,
            "page_title": title,
            "main_text": text,
            "image": image
        })
    return articles
# Exemple d'utilisation :
rss_url = "https://www.lemonde.fr/rss/en_continu.xml"  # RSS du Monde, chaque fichier RSS d’un média correspond à une catégorie ou à un type d’actualité
articles = get_rss_articles(rss_url)
for a in articles[:5]:
    print(a)
````
